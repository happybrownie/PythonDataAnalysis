#**全球温室气体排放变化 **  


```python
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt

Gas=pd.read_csv('/home/mw/input/greenhousegas6025/Greenhouse_Gas.csv')#导入数据
Gas=Gas.drop(Gas.columns[0:11],axis=1)
df=Gas[1089:]
df=df[df['Gas_Type']=='Greenhouse gas']#选出world温室气体排放数据
data=df.drop(df.columns[0:2],axis=1)
sums=data.sum()#计算温室气体排放量总和

#绘图
fig = plt.figure(figsize=(10,6))
plt.title('Changes in global greenhouse gas emissions from 2010 to 2021')
plt.plot(sums, # y轴数据
         linestyle = '-', # 折线类型
         linewidth = 2, # 折线宽度
         color = '#ff9999', # 折线颜色
         marker = 'o', # 点的形状
         markersize = 6, # 点的大小
         markeredgecolor='black', # 点的边框色
         markerfacecolor='#ff9999') # 点的填充色
plt.ylabel('Greenhouse gas emissions（Unit：Million metric tons of CO2 equivalent）')
plt.xlabel('Year')
plt.show()
```

```python
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np

energy=pd.read_csv('/home/mw/input/energy3938/Energy_Transition.csv')#导入数据
energy=energy.drop(energy.columns[0:4],axis=1)
energy=energy.drop(energy.columns[3:8],axis=1)
energy=energy.drop(energy.columns[1],axis=1)
df=energy[:20]
df=df[df['Indicator']=='Electricity Generation']
df=df.drop(df.columns[0],axis=1)

# 使用 pivot_table() 函数将数据透视
df=pd.pivot_table(df,values=df.columns[1:],index=['Energy_Type'],aggfunc='sum')
# 分别获取 Total Non-Renewable 和 Total Renewable 能源类型的数据
non_renewable=df.loc['Total Non-Renewable'].values[0:]
renewable=df.loc['Total Renewable'].values[0:]
# 获取年份信息
years=list(df.columns[0:])
# 将 Total Non-Renewable 和 Total Renewable 的数据相加
total=np.add(non_renewable,renewable)
df_sum=df.sum(axis=0)

# 计算每年Total Renewable和Total Non-Renewable占比
df_renewable_ratio=df.loc["Total Renewable",:]/df_sum
df_non_renewable_ratio=df.loc["Total Non-Renewable",:]/df_sum

# 绘制细分条形图和折线图
fig,ax1=plt.subplots(figsize=(20,6))
ax2=ax1.twinx()

ax1.bar(df.columns,df_renewable_ratio,color="#00703c",label="Total Renewable")
ax1.bar(df.columns,df_non_renewable_ratio,color="#c23b22",bottom=df_renewable_ratio,label="Total Non-Renewable")
ax1.set_xlabel("Year")
ax1.set_ylabel("Ratio")
ax1.legend(loc='upper left')

ax2.plot(years,total,label='Total Energy Consumption',linestyle='-',marker='o',color='#002147')
ax2.set_ylabel('Total Energy Consumption /Unit:Gigawatt-hours (GWh)')
ax2.set_ylim(0,13000000)
ax2.legend(loc='upper right')

plt.title('Electricity Generation',loc='left')
plt.show()
```

1 千兆瓦时 (GWh) = 1000000 千瓦时 (kWh)  
电力10000千瓦时=1.229吨标煤

```python
# 利用ADF检验时间序列数据是否平稳
def CheckStationarity(y, lags_plots=48):
    y = pd.Series(y)    # 构建带标签的的一维数组
    fig = plt.figure(figsize=(24,8))
    ax1 = plt.subplot2grid((3, 3), (0, 0))
    ax2 = plt.subplot2grid((3, 3), (1, 0))
    ax3 = plt.subplot2grid((3, 3), (2, 0))
    y.plot(ax=ax1)
    ax1.set_title('Temperature Variation')
    plot_acf(y, lags=lags_plots, zero=False, ax=ax2);   # 绘制自相关函数ACF图
    plot_pacf(y, lags=lags_plots, zero=False, ax=ax3);   # 绘制偏自相关函数PACF图
    plt.tight_layout()  # 自动调整各子图的间距
    plt.show()
    
    # 计算
    adfinput = adfuller(y)
    # 调整输出格式
    adfresult = pd.Series(adfinput[0:4], index=['Test Statistic', 'P-value', 'Lags Used', 'Number of Observations Used'])
    adfresult = round(adfresult, 3)
    for key, value in adfinput[4].items():
        adfresult["Critical Value (%s)" % key] = value.round(3)
    # 输出
    print('Results of the Test:')
    print(adfresult)
    if adfresult[0].round(2) < adfresult[5].round(2):
        print('\nThe Test Statistics is lower than the Critical Value of 5%.\nThe series seems to be stationary')
    else:
        print("\nThe Test Statistics is higher than the Critical Value of 5%.\nThe series isn't stationary")

CheckStationarity(df['LandAverageTemperature'])  
```

3. 建立SARIMA模型

3.1 通过网格搜索并依据AIC准则寻找模型最优参数

```python
p = d = q = range(0, 2)
pdq = list(itertools.product(p, d, q))
pdq_x_PDQs = [(x[0], x[1], x[2], 12) for x in list(itertools.product(p, d, q))]
a = []
b = []
c = []
wf=pd.DataFrame()
for param in pdq:
    for seasonal_param in pdq_x_PDQs:
        try:
            mod = sm.tsa.statespace.SARIMAX(df['LandAverageTemperature'],order=param,seasonal_order=seasonal_param,enforce_stationarity=False,enforce_invertibility=False)
            results = mod.fit()
            print('ARIMA{}x{} - AIC:{}'.format(param, seasonal_param, results.aic))
            a.append(param)
            b.append(seasonal_param)
            c.append(results.aic)
        except:
            continue
wf['pdq']=a
wf['pdq_x_PDQs']=b
wf['aic']=c
print(wf[wf['aic']==wf['aic'].min()])
```

3.2 检验模型精确度的方法

```python
# 3.2.1 计算RSME值
def RMSE(y_true, y_pred):
    '''均方根误差RMSE通常用于评估预测模型的准确性，值越小越准确'''
    return sqrt(mean_squared_error(y_true,y_pred))
# 3.2.2 绘图展示实际值、预测值、残差之间的关系
def plot_error(data):
    plt.figure(figsize=(24,8))
    ax1 = plt.subplot2grid((3, 3), (0, 0))
    ax2 = plt.subplot2grid((3, 3), (1, 0))
    ax3 = plt.subplot2grid((3, 3), (2, 0))
    # 同时绘制验证样本集的实际值和预测值
    ax1.plot(data.iloc[:, 0:2])
    ax1.legend(['Actual', 'Pred'])
    ax1.set_title('Actual and Predicted Values for Validation Set')
    ## 绘制残差的Q-Q图，用于评估误差是否服从正态分布
    sm.graphics.qqplot(data.loc[:, 'Diff'], line='r', ax=ax2)
    # 绘制残差的自相关图
    plot_acf(data.iloc[:, 2], lags=(len(data.loc[:, 'Diff']) - 1), zero=False, ax=ax3)
    plt.tight_layout()
    plt.show()
```
